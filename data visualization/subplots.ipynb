```python
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
import numpy as np
```

```python
def get_rand():
    return ((np.random.rand(10) * 100)) + 1
```

```python
r1 = get_rand()
r2 = get_rand()
r3 = get_rand()
r4 = get_rand()
```

```python
r1, r2, r3, r4
```

```python
#figsize = absolute size of space for plots
plt.figure(figsize=(8,4))
plt.subplots_adjust(hspace=.25)

#plt.subplot(nrows, ncols, index)
plt.subplot(2, 2, 1)
plt.title('r1')
plt.plot(r1, color='b', linestyle='-', linewidth=1)
plt.subplot(2, 2, 2)
plt.title('r2')

plt.subplots_adjust(hspace=.5)
plt.plot(r2, color='b', linestyle='-', linewidth=1)
plt.subplot(2, 2, 3)
plt.title('r3')
plt.plot(r3, color='b', linestyle='-', linewidth=1)
plt.subplot(2, 2, 4)
plt.title('r4')
plt.plot(r4, color='b', linestyle='-', linewidth=1)
```

```python
plt.figure(figsize=(6,6))
plt.subplots_adjust(hspace=.25)

plt.subplot(2, 1, 1)
plt.title('random nums')
plt.plot(r1, color='b', linestyle='-', linewidth=1)
plt.plot(r2, color='g', linestyle='-', linewidth=1)
```
