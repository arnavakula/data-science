```python
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
```

```python
data = load_boston()
x = data.data
y = data.target
```

```python
#histogram
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.hist(y, bins=50)
plt.xlabel('price in 1000s USD')
plt.ylabel('number of houses')
```

```python
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.scatter(x[:,5], y)
plt.xlabel('price')
plt.ylabel('number of houses')
```

```python
flight_data = sns.load_dataset('flights')
flight_data.head()
```

```python
flight_data = flight_data.pivot('month', 'year', 'passengers') #rearrange data
flight_data
```

```python
sns.heatmap(flight_data)
```

```python
job_data = ['40', '20', '17', '8', '5', '10']
labels = ['IT', 'Finance', 'Marketing', 'Admin', 'HR', 'Operations']

#explode = distinguished one
explode = (0, 0.05, 0, 0, 0, 0)

plt.pie(job_data, labels=labels, explode=explode)
plt.show()
```
