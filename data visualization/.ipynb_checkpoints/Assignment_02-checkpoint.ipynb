#### 1: List the causes of death and percentile in Ohio state in the year 2012

```python
#import required libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#cause of deaths
causes = ['Chronic Disease', 'Unintetional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis', 'Other']
#percentile
percentiles = [62, 5, 4, 2, 1 ,26]
```

#### 2: Draw a pie chart using the two data points

```python
#set pie chart properties
explode = (0.025, 0, 0, 0, 0, 0)
plt.figure(figsize=(12, 12))
plt.pie(percentiles, labels=causes, autopct='%1.1f%%', startangle=70, explode=explode)
plt.axis('equal')
plt.title('Ohio State Leading Causes of Death - 2012')
plt.show()
```
