```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
random_nums = np.random.rand(10)
random_nums
```

```python
style.use('ggplot')

#set parameters (data, color, label, width)
plt.plot(random_nums, 'g', label='random_nums line')

plt.xlabel('Index')
plt.ylabel('Numbers')

plt.title('Practice Plot')
```
