```python
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline

import seaborn as sns
```

```python
data = pd.read_csv('auto_data.csv')
```

```python
data.head()
```

```python
#graph mpg (continuous data --> histogram)
style.use('ggplot')
plt.figure(figsize=(7, 7))

plt.hist(data['mpg'], bins=30)
plt.xlabel('mpg')
plt.ylabel('freq')
```

```python
#graph weight
style.use('ggplot')
plt.figure(figsize=(7, 7))

plt.hist(data['weight'], bins=50)
plt.xlabel('weight.')
plt.ylabel('freq')
```

```python
#graph using seaborn
def origin(num):
    if num == 1:
        return 'USA'
    elif num == 2:
        return 'Europe'
    else: 
        return 'Asia'

data['origin'] = data['origin'].apply(origin)
```

```python
data.head(30)
```

```python
sns.pairplot(data[['mpg', 'weight', 'origin']], hue='origin', height=4)
```
